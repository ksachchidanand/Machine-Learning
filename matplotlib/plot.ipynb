```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

### Basic graphs

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

#plt.plot(x, y, label='2x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='blue', linestyle='--')

#use shorthand notation
#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')

## line number two

x2 = np.arange(0, 4.5, 0.5)
print(x2)

plt.plot(x2, x2**2, 'rx--', label='x^2')


plt.title('our first graph', fontdict={'fontname': 'Comic Sans MS'})
plt.xlabel('X Axis', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Y Axis')

plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 8, 10])

plt.legend()

plt.show()
```
